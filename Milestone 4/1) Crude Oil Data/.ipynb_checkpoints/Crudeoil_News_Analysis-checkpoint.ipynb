```python
import pandas as pd
import numpy as np
from datetime import timedelta
```

```python
oilprice = pd.read_csv('crudeoil_preprocessed.csv', thousands=',')
```

```python
print(oilprice.info())
```

```python
len(oilprice)
```

```python
oilprice['trend'] = ''
```

```python
# Change Date to date format and reorder dataframe

oilprice['Date'] = pd.to_datetime(oilprice['Date'])
oilprice.sort_values(by=['Date'], ascending=False, inplace=True)
oilprice = oilprice.reset_index(drop=True)
oilprice.dropna(inplace=True)
```

```python
oilprice.head(10)
```

```python
for i in range(0, len(oilprice)):  # from row 1 untl end of the row
    if oilprice['Close_Price'][i] > oilprice['Close_Price'][i+1]:
        oilprice['trend'][i] = 'upward'
    elif oilprice['Close_Price'][i] == oilprice['Close_Price'][i+1]:
        oilprice['trend'][i] = 'maintain'
    else:
        oilprice['trend'][i] = 'downward'
```

```python
oilprice.head()
```

```python
def get_day(x):
    return x.day_name()


def get_day_no(x):
    return x.weekday() + 1


def get_month(x):
    return x.month_name()
```

```python
oilprice['day'] = oilprice['Date'].apply(get_day)
oilprice['day_no'] = oilprice['Date'].apply(get_day_no)
oilprice['month'] = oilprice['Date'].apply(get_month)
oilprice['year'] = oilprice['Date'].dt.year
```

```python
oilprice = oilprice[['year', 'month', 'Date', 'day', 'day_no', 
     'Open_Price', 'Low_Price', 'High_Price', 'Close_Price', 
     'trend', 'Open_Equals_Closing', 'All_Day_Maintain', 'Closing_Low', 'Closing_High']]
```

```python
oilprice.head()
```

```python
oilprice.to_csv('oilprice_trend.csv', index=False)
```

```python
for i in range(0,len(oilprice)):
    if np.isnan(oilprice['Close_Price'][i]):
        oilprice['Close_Price'][i] = np.mean([oilprice['Close_Price'][i-1],oilprice['Close_Price'][i+1]])
    if np.isnan(oilprice['Open_Price'][i]):
        oilprice['Open_Price'][i] = np.mean([oilprice['Open_Price'][i-1],oilprice['Open_Price'][i+1]])
    if np.isnan(oilprice['Low_Price'][i]):
        oilprice['Low_Price'][i] = np.mean([oilprice['Low_Price'][i-1],oilprice['Low_Price'][i+1]])
    if np.isnan(oilprice['High_Price'][i]):
        oilprice['High_Price'][i] = np.mean([oilprice['High_Price'][i-1],oilprice['High_Price'][i+1]])
```

```python
# Call news data csv

get_polarity_mean = pd.read_csv('polarity_mean.csv')
get_polarity_mean['Date'] = pd.to_datetime(get_polarity_mean['Date'])
```

```python
merge = pd.merge(get_polarity_mean,
                 oilprice[['Date','trend']],
                 how='left',
                 left_on=['Date'],
                right_on=['Date'])
```

```python
merge = merge.reset_index(drop=True)
merge.dropna(how='any', inplace=True)
```
